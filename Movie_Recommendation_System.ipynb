# Data-Preprocessing

```python
# Extracting the important features for our data manipulation
data = df[['budget','genres','movieid','keywords','release_date','revenue','title','vote_average','vote_count']]
```

```python
data.head()
```

```python
# droping the null values
data.dropna(inplace = True)
```

```python
data.duplicated().sum()
```

Here we are importing the "ast.literal_eval(obj)",because it will convert the string list into the normal list.

```python
# here we can see that our list is in the string format, and we are applying the ast libraries to do the same
data.iloc[0].genres
```

```python
import ast
```

```python
# Here we have made this function to fetch the name key value extraction from the genres features
def convert(obj):
  list1 =[]
  for i in ast.literal_eval(obj):
    list1.append(i['name'])
  return list1
```

```python
data['genres']= data['genres'].apply(convert)
data['keywords']= data['keywords'].apply(convert)
```

```python
data.head()
```

```python
# Concatinating the two features to form one single features
data['genres']=data['genres']+data['keywords']
```

```python
data.head()
```

```python
# Here we are droping the keywords column because it is already got concatinated in genres column
data.drop(['keywords'],axis=1,inplace=True)
```

#Stemming

```python
# Dowloading the required libraries
import nltk
nltk.download("stopwords")
```

```python
X=data['genres']
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
tag =[]
for i in range(0, len(X)):
  some_digit = X.to_numpy()[i]
  review = [ps.stem(i) for i in some_digit if not i in stopwords.words('english')]  #Removing the stopwords such as Stop words are a set of commonly used words in a language like grammatical words.
  review = ' '.join(review)
  tag.append(review)
```

```python
tag
```

#Count-Vectorizer

```python
# Importing the count-vectorizer->It is used to convert the text into the numerical vector
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer() # 
X_cv = vectorizer.fit_transform(tag).toarray()
features_name=vectorizer.get_feature_names_out()
```

```python
X_df=pd.DataFrame(X_cv,columns=features_name)
X_df
```

# Cosine Similarity

```python
#Cosine similarity- Here we are introducing cosine similarity to compare the similarity between the records of the features.it wrks on the correlation cosine angle.
from sklearn.metrics.pairwise import cosine_similarity
similarity =cosine_similarity(X_cv)
similarity
```

```python
similarity_df=pd.DataFrame(similarity,columns=data['title'],index=data['title'])
similarity_df
```
